```python
import json
import shelve
from HelperClass import InvertedIndex
import pandas as pd
```

```python
invIndex = InvertedIndex()
invIndex.loadAll('DevShelve')
```

```python
import json
termDict = {}
for term in invIndex['pos']:
    docList = list(invIndex['pos'][term])
    dfCount = len(invIndex['pos'][term])
    tDict = {'df':dfCount, 'tf':{}, 'docIds':list()}
    for docId in invIndex['pos'][term]:
        tDict['tf'].update({docId:len(invIndex['pos'][term][docId])})
    tDict['docIds'] = docList
    termDict[term] = json.dumps(tDict)
```

```python
with open('Postings.txt', "w") as fp:
    for term in termDict:
        fp.write(f'{term}-{termDict[term]}\n')
```

```python
fpDict = {}
import os
with open('Postings.txt', "r+") as fp:
    while not (fp.tell() == os.fstat(fp.fileno()).st_size):
        fpPos = fp.tell()
        line = fp.readline().split('-')
        term = line[0]
        fpDict[term] = fpPos
```

```python
def _df_from_dict(dictObj:dict, toInt = False):
    df = pd.DataFrame.from_dict(dictObj, orient='index')
    df = df.astype('int64')
    df.columns = ['fp']
    return df

df = _df_from_dict(fpDict)
df['fp']['hello']
# df = df.to_hdf(f'DevHDF5/PostFp.hdf5', key='fp')
```

```python
def readLine(term:str, fileName:str, df):
    with open(fileName, "r+") as fp:
        fp.seek(df['fp'][term])
        return fp.readline().strip().split('-')[1]

termInfo = readLine('hello', 'Data/Postings.txt',df)
termInfo = json.loads(termInfo)
print(termInfo)
```
